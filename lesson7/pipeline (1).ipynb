```python
# Numbers

numbers = [-5, -3, 0, 2, 4, 6, 9]

def filter_positive(numbers):
    return [n for n in numbers if n > 0]

def square_numbers(numbers):
    return [n ** 2 for n in numbers]

def sum_numbers(numbers):
    return sum(numbers)
  
# Pipeline

steps = [filter_positive, square_numbers, sum_numbers]

result = numbers
for step in steps:
    result = step(result)
    print(f"{step.__name__}: {result}")
```

```python
# File

filename = "My Document File"

def to_lowercase(filename):
    return filename.lower()

def replace_spaces(filename):
    return filename.replace(" ", "_")

def add_extension(filename):
    return filename + ".txt"

# Pipeline
steps = [to_lowercase, replace_spaces, add_extension]

result = filename
for step in steps:
    result = step(result)
    print(f"{step.__name__}: {result}")
```

```python
# Password

password = "hello123"

def check_length(password):
    if len(password) >= 8:
        print("check_length: ✅ uzunlik yetarli")
        return password
    else:
        print("check_length: ❌ kamida 8 ta belgi kerak")
        return None

def check_uppercase(password):
    if password is None:
        return None
    if any(c.isupper() for c in password):
        print("check_uppercase: ✅ katta harf bor")
        return password
    else:
        print("check_uppercase: ❌ katta harf yo'q")
        return None

def check_digit(password):
    if password is None:
        return None
    if any(c.isdigit() for c in password):
        print("check_digit: ✅ raqam bor")
        return password
    else:
        print("check_digit: ❌ raqam yo'q")
        return None

# Pipeline
steps = [check_length, check_uppercase, check_digit]

result = password
for step in steps:
    result = step(result)

print("\n✅ Parol kuchli!" if result else "\n❌ Parol kuchsiz!")
```

```python
# Enail

email = "hello@gmail.com"

def check_at_symbol(email):
    if email is None:
        return None
    if "@" in email:
        print("check_at_symbol: ✅ @ belgisi bor")
        return email
    else:
        print("check_at_symbol: ❌ @ belgisi yo'q")
        return None

def check_domain(email):
    if email is None:
        return None
    if any(email.endswith(d) for d in [".com", ".org", ".net", ".uz"]):
        print("check_domain: ✅ domain to'g'ri")
        return email
    else:
        print("check_domain: ❌ domain noto'g'ri")
        return None

def check_length(email):
    if email is None:
        return None
    if len(email) >= 5:
        print("check_length: ✅ uzunlik yetarli")
        return email
    else:
        print("check_length: ❌ email juda qisqa")
        return None

# Pipeline
steps = [check_at_symbol, check_domain, check_length]

result = email
for step in steps:
    result = step(result)

print("\n✅ Email to'g'ri!" if result else "\n❌ Email noto'g'ri!")
```
