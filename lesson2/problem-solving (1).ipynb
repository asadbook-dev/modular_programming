```python
# Library — kitob qo'shish va qidirish
```

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append({"title": title, "author": author})
        print(f"'{title}' qo'shildi!")

    def search(self, title):
        for book in self.books:
            if book["title"].lower() == title.lower():
                return f"Topildi: '{book['title']}' — {book['author']}"
        return "Kitob topilmadi."

lib = Library()
lib.add_book("O'tkan kunlar", "Abdulla Qodiriy")
lib.add_book("Mehrobdan chayon", "Abdulla Qodiriy")
print(lib.search("O'tkan kunlar"))  # Topildi: 'O'tkan kunlar' — Abdulla Qodiriy
```

```python
# Car — yoqilg'i va harakat
```

```python
class Car:
    def __init__(self, brand, fuel=50):
        self.brand = brand
        self.fuel = fuel
        self.km = 0

    def drive(self, distance):
        if self.fuel >= distance * 0.1:
            self.km += distance
            self.fuel -= distance * 0.1
            print(f"{distance} km yurdi. Qolgan yoqilg'i: {self.fuel:.1f}L")
        else:
            print("Yoqilg'i yetarli emas!")

    def refuel(self, amount):
        self.fuel += amount
        print(f"Yoqilg'i qo'shildi. Jami: {self.fuel}L")

car = Car("Cobalt")
car.drive(100)     # 10L sarfladi
car.refuel(20)     # 60L bo'ldi
car.drive(400)     # Yoqilg'i yetarli emas!
```

```python
# Student — baho qo'shish va attestatsiya
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = {}

    def add_grade(self, subject, grade):
        self.grades[subject] = grade

    def average(self):
        if not self.grades:
            return 0
        return sum(self.grades.values()) / len(self.grades)

    def passed(self):
        return "O'tdi ✓" if self.average() >= 60 else "Qoldi ✗"

s = Student("Jasur")
s.add_grade("Matematika", 85)
s.add_grade("Ingliz tili", 72)
s.add_grade("Fizika", 55)
print(f"O'rtacha: {s.average():.1f}")  # 70.7
print(s.passed())                       # O'tdi ✓
```
