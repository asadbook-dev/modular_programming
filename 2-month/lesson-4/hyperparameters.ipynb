```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



df=pd.read_csv('StudentPerformanceFactors.csv')
```

```python
df.info() 
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.dtypes
```

```python
encoder=LabelEncoder()

for col in df.columns:
    df[col]=encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
x=df.drop('Exam_Score',axis=1)
y=df['Exam_Score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Manual Search

```python
from sklearn.ensemble import RandomForestRegressor

rf_manual_model=RandomForestRegressor(n_estimators=100,max_depth=5,min_samples_split=2,min_samples_leaf=1,random_state=42)

rf_manual_model.fit(x_train,y_train)
```

```python
y_predict_manual=rf_manual_model.predict(x_test)
r2_manual = r2_score(y_test, y_predict_manual)
print(f"R2 Score for Manual Search: {r2_manual:.4f}")
```

```python
r2_manual
```

# Grid Search

```python
# Grid Search for Hyperparameter Tuning

# import GridSearchCV from sklearn.model_selection
from sklearn.model_selection import GridSearchCV

# Creating a dictionary of hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
}



# GridSearch parameters =>
# estimator: The model to be tuned (RandomForestRegressor in this case).
# param_grid: The dictionary of hyperparameters to search over.
# cv: The number of cross-validation folds (5 in this case).
# scoring: The metric to optimize (r2 in this case).
# n_jobs: The number of parallel jobs to run (-1 means use all processors).
grid_search=GridSearchCV(estimator=RandomForestRegressor(random_state=42), param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
grid_search
```

```python
print("Grid Search Best Params:", grid_search.best_params_)
```

```python
# Training

# with found best parameters, we can create a new RandomForestRegressor model and fit it to the training data.
rf_grid_model=RandomForestRegressor(**grid_search.best_params_, random_state=42)
rf_grid_model.fit(x_train, y_train)
```

```python
# Prediction

y_predict_grid=rf_grid_model.predict(x_test)
r2_grid=r2_score(y_test, y_predict_grid)
print(f"R2 Score for Grid Search: {r2_grid:.4f}")
```

# Random Search

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 5, 10],
}
```

```python
random_search=RandomizedSearchCV(estimator=RandomForestRegressor(random_state=42), param_distributions=param_dist, n_iter=5, cv=5, scoring="r2", n_jobs=-1, random_state=42)
random_search.fit(x_train, y_train)
print("Random Search Best Params:", random_search.best_params_)
```

```python
print(random_search.best_estimator_)
```

```python
print(random_search.best_score_)
```

```python
print(random_search.best_index_)
```

```python
print(random_search.feature_names_in_)
```

```python
# Training with best hyperparameters from RandomizedSearchCV
rf_random_model=RandomForestRegressor(**random_search.best_params_, random_state=42)
rf_random_model.fit(x_train, y_train)
```

```python
# Prediction

y_predict_random=random_search.predict(x_test)
r2_random=r2_score(y_test, y_predict_random)
print(f"R2 Score for Random Search: {r2_random:.4f}")
```

```python
print(random_search.best_params_)
print(random_search.best_score_)
print(f"R2 Score for Random Search: {r2_random:.4f}")

```

# Byesian Optimization

```python
# we need (scikit-optimize library)

from skopt import BayesSearchCV

param_space = {
    "n_estimators": (20, 150), # from 20 to 150 => 20, 21, 22, 23, ..., 149, 150
    "max_depth": (3, 15) # from 3 to 15 => 3, 4, ..., 14, 15
}

```

```python
bayes_search=BayesSearchCV(estimator=RandomForestRegressor(random_state=42), search_spaces=param_space, n_iter=10, cv=5, random_state=42, n_jobs=-1)
bayes_search.fit(x_train, y_train)
```

```python
print(bayes_search.best_params_)
```

```python
print(bayes_search.best_score_)
```

```python
# Training

rf_bayes=RandomForestRegressor(**bayes_search.best_params_)
rf_bayes.fit(x_train, y_train)
```

```python
# prediction

y_predict_bayes=rf_bayes.predict(x_test)
r2_bayes=r2_score(y_test, y_predict_bayes)
print(f"R2 Score for Bayesian Optimization: {r2_bayes:.4f}")
```

```python
print(bayes_search.best_params_)
print(bayes_search.best_score_)
print(f"R2 Score for Random Search: {r2_bayes:.4f}")

```

```python
import optuna
from sklearn.model_selection import cross_val_score 

def objective_func(trial):
    n_estimators=trial.suggest_int("n_estimators", 50, 100)
    max_depth=trial.suggest_int("max_depth", 2, 10)

    clf = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=42)

    return cross_val_score(clf, x_train, y_train, cv=5).mean()


# direction="maximize" => beacause we are using r2_score we need higher score
# if we need lower score like mae, mse direction="minimize"
study=optuna.create_study(direction="maximize")

# we are calling objective_func 10 times because n_trials=10
study.optimize(objective_func, n_trials=10)
```

```python
print("Optuna Best Params:", study.best_params)
```

```python
# Training

rf_optuna=RandomForestRegressor(**study.best_params, random_state=42)
rf_optuna.fit(x_train, y_train)
```

```python
#  Prediction

y_predict_optuna=rf_optuna.predict(x_test)
r2_optuna=r2_score(y_test, y_predict_optuna)
print(f"R2 Score for Optuna: {r2_optuna:.4f}")
```

```python
print(study.best_params)
print(study.best_value)
print(f"R2 Score for Optuna: {r2_optuna:.4f}")

```
