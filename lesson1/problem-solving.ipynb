```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Yetarli mablag' yo'q!")
```

```python
# 1. Account yaratish
account = BankAccount("G'ani", 80000)

# 2. Deposit
account.deposit(30000)
print(account.balance)  # 110000

# 3. Withdraw
account.withdraw(20000)
print(account.balance)  # 90000
```

```python
# StudentGrades nomli class yarating va undan talabalarning ismi, baxolar ro'yxati va shunga ko'ra ularning ichida eng past va eng yuqori ball olgan talaba va o'rtacha ballni qaytarsin
```

```python
class StudentGrades:
    def __init__(self):
        self.students = {}

    def add_student(self, name, grades):
        self.students[name] = grades

    def average(self):
        all_grades = [g for grades in self.students.values() for g in grades]
        return round(sum(all_grades) / len(all_grades), 1)

    def highest(self):
        return max(self.students, key=lambda n: max(self.students[n]))

    def lowest(self):
        return min(self.students, key=lambda n: min(self.students[n]))
```

```python
sg = StudentGrades()
sg.add_student("Aziz",   [88, 92, 76, 95])
sg.add_student("Malika", [70, 65, 80, 72])
sg.add_student("Jasur",  [95, 98, 100, 92])

print("O'rtacha ball:", sg.average())   # 85.2
print("Eng yuqori:",    sg.highest())   # Jasur
print("Eng past:",      sg.lowest())    # Malika
```
